```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Configuring plots

sns.set_style('darkgrid')
sns.set(rc={'figure.figsize':(1.5*11.7,1.5*8.27)})
```

```python
df_payments_monthly = pd.read_csv('../data/monthlyPayment.csv')
```

```python
df_payments_monthly.drop('ClientId', axis=1, inplace=True)
```

```python
MRR = df_payments_monthly.sum()
MRR = MRR.drop('January/2020')
```

```python
df_mrr = MRR.to_frame()
df_mrr['Month'] = df_mrr.index
df_mrr.reset_index(drop=True, inplace=True)
df_mrr.rename({0: 'MRR'}, axis=1, inplace=True)
```

```python
ax = sns.pointplot(x="Month", y="MRR", data=df_mrr, markers=["o"], linestyles=["-"])
ax.set_xticklabels(list(df_mrr.Month), rotation=90);
ax.set_xlabel("Months",fontsize=20);
ax.set_ylabel("MRR (BRL)",fontsize=20);
ax.tick_params(labelsize=20);
```

In the above plot we can see that the MRR value was stable up to June/2019 with no major trend or seasonality apparent. After this we see a spike in August/2019 followed by a strong dip that lowered the MRR by less than 20% of the value before June/2019.

Further analyzes considering the contributions of each Market Segment and each Region might show more insights for the major dip.
